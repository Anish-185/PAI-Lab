```python
import numpy as np

grid_size = 3
gamma = 0.9

V = np.zeros((grid_size, grid_size))
policy = np.zeros((grid_size, grid_size), dtype=int)

actions = [(0,1),(1,0),(0,-1),(-1,0)]
action_names = ["RIGHT", "DOWN", "LEFT", "UP"]

def get_next(i,j,a):
    ni = i + a[0]
    nj = j + a[1]

    if ni < 0 or ni >= grid_size or nj < 0 or nj >= grid_size:
        return i,j
    return ni,nj

# policy iteration
for _ in range(5):

    # Policy Evaluation
    for _ in range(5):
        for i in range(grid_size):
            for j in range(grid_size):
                a = actions[policy[i][j]]
                ni,nj = get_next(i,j,a)
                V[i][j] = -1 + gamma * V[ni][nj]

    # Policy Improvement
    for i in range(grid_size):
        for j in range(grid_size):
            values = []

            for a in actions:
                ni,nj = get_next(i,j,a)
                values.append(-1 + gamma * V[ni][nj])

            policy[i][j] = np.argmax(values)

# PRINT OUTPUT LIKE YOUR IMAGE
print("\nPolicy Iteration Results:")
for i in range(grid_size):
    for j in range(grid_size):
        action = action_names[policy[i][j]]
        print(f"State: ({i}, {j}), Action: {action}, Value: {round(V[i][j], 2)}")
```
