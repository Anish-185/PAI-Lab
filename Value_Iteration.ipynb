```python

import numpy as np
```

```python
grid_size=3
gamma=0.9
```

```python
V = np.zeros((int(grid_size),int(grid_size)))
```

```python
actions=[(0,1),(1,0),(0,-1),(-1,0)]

def get_next(i,j,a):
    ni=i+a[0]
    nj=j+a[1]

    if ni < 0 or ni >= grid_size or nj < 0 or nj >= grid_size:
        return i,j
    return ni,nj
```

```python
for _ in range(10):
  new_V =np.copy(V)

  for i in range(int(grid_size)):
    for j in range(int(grid_size)):
      values=[]

      for a in actions:
        ni,nj=get_next(i,j,a)
        values.append(-1+gamma * V[ni][nj])

      new_V[i][j]=max(values)

  V=new_V
```

```python
print("Values  iteration results = ")
for i in range(int(grid_size)):
  for j in range(int(grid_size)):
    print(f"State: ({i}, {j}), Value: {round(V[i][j], 2)}")
```
