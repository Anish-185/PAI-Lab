```python
!pip install pgmpy

from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination
from pgmpy.factors.discrete import TabularCPD

# 1. Create the model (structure)
model = DiscreteBayesianNetwork([
    ('Income', 'Default'),
    ('Credit', 'Default')
])

# 2. Define probabilities

# Income: High, Low
cpd_income = TabularCPD(
    variable='Income',
    variable_card=2,
    values=[[0.6], [0.4]]  # 60% High, 40% Low
)

# Credit: Good, Poor
cpd_credit = TabularCPD(
    variable='Credit',
    variable_card=2,
    values=[[0.7], [0.3]]  # 70% Good, 30% Poor
)

# Default Risk: Low, High
cpd_default = TabularCPD(
    variable='Default',
    variable_card=2,
    values=[
        [0.9, 0.6, 0.5, 0.1],  # Low Risk
        [0.1, 0.4, 0.5, 0.9]   # High Risk
    ],
    evidence=['Income', 'Credit'],
    evidence_card=[2, 2]
)

# 3. Add everything to model
model.add_cpds(cpd_income, cpd_credit, cpd_default)

# 4. Check model
print("Is model correct?", model.check_model())

# 5. Ask a question (prediction)
inference = VariableElimination(model)

result = inference.query(
    variables=['Default'],
    evidence={'Income': 1, 'Credit': 1}  # Low income, Poor credit
)

print("\nDefault Risk Prediction:")
print(result)
```
