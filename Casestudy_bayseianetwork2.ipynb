```python
!pip install pgmpy
from pgmpy.models import DiscreteBayesianNetwork
from pgmpy.inference import VariableElimination
from pgmpy.factors.discrete import TabularCPD

# 1. Create model
model = DiscreteBayesianNetwork([('Weather', 'Picnic')])

# 2. Define probabilities

# Weather: Sunny(0), Rainy(1)
cpd_weather = TabularCPD(
    variable='Weather',
    variable_card=2,
    values=[[0.7], [0.3]]  # 70% Sunny, 30% Rainy
)

# Picnic: Yes(0), No(1)
cpd_picnic = TabularCPD(
    variable='Picnic',
    variable_card=2,
    values=[
        [0.9, 0.2],  # Picnic = Yes
        [0.1, 0.8]   # Picnic = No
    ],
    evidence=['Weather'],
    evidence_card=[2]
)

# 3. Add to model
model.add_cpds(cpd_weather, cpd_picnic);

# 4. Check model
print("Model correct?", model.check_model())

# 5. Inference
inference = VariableElimination(model)

# Case 1: If weather is Sunny (0)
result1 = inference.query(variables=['Picnic'], evidence={'Weather': 0})
print("\nIf Sunny:")
print(result1)

# Case 2: If weather is Rainy (1)
result2 = inference.query(variables=['Picnic'], evidence={'Weather': 1})
print("\nIf Rainy:")
print(result2)
```
